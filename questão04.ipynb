Quest√£o 04 - Letra A

```python
from PIL import Image
import matplotlib.pyplot as plt

def main():
    b = 50
    c = 1/10
    imagem = Image.open("Ativ.Pratica.01/lena_gray.bmp")

    width, height = imagem.size
    
    x = [0 for _ in range(width)]

    for i in range(width):
        for j in range(height):
            imagem.putpixel((i, j), round(c * imagem.getpixel((i, j)) + b))
            x[imagem.getpixel((i, j))] += 1

    imagem.save("transf_afim_2.png")
    plt.plot(x)
    plt.show()

    
if __name__ == "__main__":
    main()
```

Quest√£o 04 - Letra B

```python
from PIL import Image
import matplotlib.pyplot as plt
import math

def main():
    c = 20
    imagem = Image.open("Ativ.Pratica.01/lena_gray.bmp")

    width, height = imagem.size
    
    x = [0 for _ in range(width)]

    for i in range(width):
        for j in range(height):
            imagem.putpixel((i, j), int(c * math.log(imagem.getpixel((i, j)) + 1, 2)))
            x[imagem.getpixel((i, j))] += 1

    imagem.save("transf_logaritmica_2.png")
    plt.plot(x)
    plt.show()

    
if __name__ == "__main__":
    main()
```

Quest√£o 04 - Letra C

```python
from PIL import Image
import matplotlib.pyplot as plt
import math

def main():
    c = 1 / math.exp(200)
    imagem = Image.open("Ativ.Pratica.01/lena_gray.bmp")

    width, height = imagem.size
    
    x = [0 for _ in range(width)]

    for i in range(width):
        for j in range(height):
            imagem.putpixel((i, j), round(c * (math.exp(imagem.getpixel((i, j)) + 1))))
            x[imagem.getpixel((i, j))] += 1

    imagem.show()
    plt.plot(x)
    plt.show()

    
if __name__ == "__main__":
    main()
```
